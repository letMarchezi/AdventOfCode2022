Day 1 - Calories Count

```python
inFile = "input.txt"
with open(inFile,'r') as i:
    lines = i.readlines()
```

```python
lines
```

```python
lines_proc = [x.split("\n",1)[0] for x in lines]
```

```python
def forloop(inputs):
    j=0
    sums=[0]
    for i in inputs:
        if i == '':
            j+=1
            sums.append(0)
        else:
            sums[j] += int(i)
    return sums     
```

```python
sort_calories = sorted(forloop(lines_proc), reverse=True)
```
